```python
file_names = {}
for link in link_to_table.keys():
    response = requests.get(url=link)
    name = link_to_table[link]
    file_name = f'блядские таблицы/{name}_hse_eda_applications_list_2024.xlsx'
    file_names[file_name] = name
    file_to_write = open(file_name, mode='wb')
    file_to_write.write(response.content)
```

```python
for table in file_names:
    xlsx_file = pd.read_excel(table, header=14, index_col=0)
    xlsx_file.drop((xlsx_file[xlsx_file['Поступление на места\nв рамках отдельной квоты'] == 'Да'].index), inplace=True)
    xlsx_file.sort_values(by=['Право поступления\nбез вступительных испытаний', 'Поступление на места\nв рамках отдельной квоты', 'Поступление на места в рамках квоты\nцелевого приема', 
                              'Поступление на места в рамках квоты \nдля лиц, имеющих особое право', 'Сумма конкурсных баллов', 'Приоритет иных мест'], 
                          ascending=[True, False, False, False, False, True], inplace=True, ignore_index=True)
    link = table.split('/')
    xlsx_file.to_excel(f'{link[0]}/formatted_{link[1]}')
    for SNILS in SNILSes:
        lat = xlsx_file.loc[xlsx_file['СНИЛС / Уникальный идентификатор'] == SNILS]
        if not lat.empty:
            lat = lat[['СНИЛС / Уникальный идентификатор', 'Право поступления\nбез вступительных испытаний', 'Сумма конкурсных баллов', 'Приоритет иных мест', 'Право поступления\nбез вступительных испытаний', 'Поступление на места\nв рамках отдельной квоты', 'Поступление на места в рамках квоты\nцелевого приема', 
                              'Поступление на места в рамках квоты \nдля лиц, имеющих особое право']]   
            lat['Образовательная программа'] = file_names[table]
            lat.reset_index(inplace=True)
            lat.set_index('Образовательная программа', inplace=True)
            lat.rename(columns={'index' : 'Положение'}, inplace=True)
            SNILSes[SNILS][f'{file_names[table]}'] = lat
xlsx_file

```

```python
for SNILS in SNILSes:
    table_of_wishes = pd.DataFrame()
    priorities = SNILSes[SNILS]
    frames = []
    for lat in priorities:
        frames.append(priorities[lat])
    table_of_wishes = pd.concat(frames)
    table_of_wishes.sort_values(by=['Приоритет иных мест'], inplace=True)
    SNILSes[SNILS] = table_of_wishes
```

```python
Table_of_priorities = pd.concat(SNILSes.values())
Table_of_priorities.reset_index(inplace=True)
Table_of_priorities.set_index('СНИЛС / Уникальный идентификатор', inplace=True)
Table_of_priorities.to_excel('THE_END_RESULT.xlsx')
```
