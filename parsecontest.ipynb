Быстрый скрипт времен ассистенства в ЛОШ, определяющий кто за день присылал решения каких-то задач. Так мы определяли посещаемость с хорошей точностью, запуская скрипт в час-два дня, потому что никто не собирается решать в такую рань задачи в контесте, если они уже не учатся

Скрипты для каждого потока (ИНФ-i) запускаются по отдельности, чтобы сохранить все данные о том, в каких группах находятся ученики.

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
```

```python
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

iterations = 14
all_data = []
try:
    url = 'https://passport.yandex.ru/auth?retpath=http%3A%2F%2Fadmin.contest.yandex.ru%2F'
    driver.get(url)

    print("Пожалуйста, вручную введите логин, пароль и код 2FA, затем нажмите Enter")
    input("Нажмите Enter после завершения аутентификации...") # без этого яндекс просто убьет нас

    for i in range(iterations):
        contest = f'https://admin.contest.yandex.ru/contests/66482/submissions?offset={i}'
        driver.get(contest)
        wait = WebDriverWait(driver, 10)
        data_elements = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, 'CompoundTable-TableBody')))
        
        all_data.append(data_elements[1].text.split('\n'))
finally:
    print('The program is done')

all_data
```

```python
attempts_inf1 = {}
attendance = {}
print(attendance)
studs = []
for lines in all_data:
    studs.extend(set(' '.join(lines[i+3].split()[:3]) for i in range(0, len(lines), 8) if (len(lines[i+3].split()) > 1 
                                                                                           and lines[i+1].split()[0]=='06.08.24')))
studs = set(studs)
for stud in studs:
    attempts_inf1[stud] = []
    attendance[stud] = 'NO'
for lines in all_data:
    for stud in studs:
        stud_attempts = set([(lines[i+1].split()[0], lines[i+2]) 
                         for i in range(0, len(lines), 8) if (' '.join(lines[i+3].split()[:3]) == stud and lines[i+1].split()[0]=='06.08.24')])
        if stud_attempts:
            attempts_inf1[stud].extend(stud_attempts)
attempts_inf1
studs
```
